```python
import yfinance as yfin


df_Eur_Usd = yfin.Ticker("EURUSD=X")

Eur_Usd_5min = df_Eur_Usd.history(period = '60d', interval = '5m')

Eur_Usd_5min.index = Eur_Usd_5min.index.tz_localize(None) # The Timezone is +1h


Eur_Usd_5min.tail()
```

```python
from datetime import date

# Get the current date
current_date = date.today().strftime("%Y-%m-%d")
df_name = r"C:\Users\berna\OneDrive\Documentos\DS_Master\Articles Forex\Data" + r"\Eur_Usd_5min_" + current_date + ".xlsx"
#print(df_name)

Eur_Usd_5min.to_excel(df_name, index=True)

print("Done")
```
